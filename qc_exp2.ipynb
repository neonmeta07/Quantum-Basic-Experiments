```python
from qiskit import QuantumCircuit
from qiskit.circuit import Gate
from qiskit_aer import AerSimulator
from math import pi
from qiskit.visualization import array_to_latex

# Initialize simulator for unitary calculations
simulator = AerSimulator()

# Function to run and display circuit unitary
def run_and_visualize_unitary(circuit, prefix="\\text{Circuit Unitary} = "):
    circuit.save_unitary()
    result = simulator.run(circuit).result()
    unitary = result.get_unitary(circuit)
    display(circuit.draw(output='mpl'))
    display(array_to_latex(unitary, prefix=prefix))

# --- Controlled Gate Identities ---

# 1. Controlled-Z Gate
qc1 = QuantumCircuit(2)
qc1.cz(0, 1)
print("Controlled-Z Gate:")
run_and_visualize_unitary(qc1)

# 2. Controlled-Z using H and CNOT
qc2 = QuantumCircuit(2)
qc2.h(1)
qc2.cx(0, 1)
qc2.h(1)
print("\nControlled-Z using H-CNOT-H:")
run_and_visualize_unitary(qc2)

# 3. Controlled-Y
qc3 = QuantumCircuit(2)
qc3.sdg(1)
qc3.cx(0, 1)
qc3.s(1)
print("\nControlled-Y:")
run_and_visualize_unitary(qc3)

# 4. Controlled-H
qc4 = QuantumCircuit(2)
qc4.ry(pi/4, 1)
qc4.cx(0, 1)
qc4.ry(-pi/4, 1)
print("\nControlled-H:")
run_and_visualize_unitary(qc4)

# --- Swap Identities ---

# 5. SWAP Gate
qc5 = QuantumCircuit(2)
qc5.swap(0, 1)
print("\nSWAP Gate:")
run_and_visualize_unitary(qc5)

# 6. SWAP using CNOTs
qc6 = QuantumCircuit(2)
qc6.cx(0, 1)
qc6.cx(1, 0)
qc6.cx(0, 1)
print("\nSWAP using CNOTs:")
run_and_visualize_unitary(qc6)

# --- Toffoli Gate Identity ---

# 7. Toffoli Gate
qc7 = QuantumCircuit(3)
qc7.ccx(0, 1, 2)
print("\nToffoli Gate:")
run_and_visualize_unitary(qc7)

# 8. Toffoli using CP and CNOT
qc8 = QuantumCircuit(3)
qc8.cp(pi/2, 1, 2)
qc8.cx(0, 1)
qc8.cp(-pi/2, 1, 2)
qc8.cx(0, 1)
qc8.cp(pi/2, 0, 2)
print("\nToffoli using CP and CNOT:")
run_and_visualize_unitary(qc8)

# --- Single-Qubit Identities ---

# 9. HXH = Z
qc9 = QuantumCircuit(1)
qc9.h(0)
qc9.x(0)
qc9.h(0)
print("\nHXH = Z:")
run_and_visualize_unitary(qc9, prefix="\\text{HXH Unitary} = ")

# 10. HZH = X
qc10 = QuantumCircuit(1)
qc10.h(0)
qc10.z(0)
qc10.h(0)
print("\nHZH = X:")
run_and_visualize_unitary(qc10, prefix="\\text{HZH Unitary} = ")

# 11. HYH = -Y
qc11 = QuantumCircuit(1)
qc11.h(0)
qc11.y(0)
qc11.h(0)
print("\nHYH = -Y:")
run_and_visualize_unitary(qc11, prefix="\\text{HYH Unitary} = ")

# --- Other Identities ---

# 12. T Gate
qc12 = QuantumCircuit(1)
qc12.t(0)
print("\nT Gate:")
run_and_visualize_unitary(qc12)

# 13. T using HTH
qc13 = QuantumCircuit(1)
qc13.h(0)
qc13.t(0)
qc13.h(0)
print("\nT using HTH:")
run_and_visualize_unitary(qc13)
```

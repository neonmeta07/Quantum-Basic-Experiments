```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator


# Function to analyze and visualize a circuit
def analyze_circuit(circuit, name="Circuit"):
    print(f"\nAnalyzing {name}")
    # Calculate metrics
    depth = circuit.depth()
    width = circuit.num_qubits
    size = circuit.size()
    print(f"Circuit Depth: {depth}")
    print(f"Circuit Width: {width} qubits")
    print(f"Circuit Size: {size} gates")
    
    # Draw circuit
    print(f"\n{name} Diagram:")
    display(circuit.draw(output='mpl'))
    


# Circuit 1: Simplified Bell-like circuit
print("=== Circuit 1 ===")
qc1 = QuantumCircuit(2, 2)  # 2 qubits, 2 classical bits for potential measurement
qc1.h(0)  # Single H gate (multiple H gates redundant)
qc1.cx(0, 1)  # Entangle qubits
qc1.h(1)  # Additional operation    
qc1.measure([0, 1], [0, 1])  # Measure for result verification
analyze_circuit(qc1, "Circuit 1")

# Circuit 2: 3-qubit circuit
print("\n=== Circuit 2 ===")
qc2 = QuantumCircuit(3, 3)  # 3 qubits, 3 classical bits
qc2.h(0)
qc2.cx(0, 1)
qc2.h(0)
qc2.h(1)
qc2.cx(1, 2)
qc2.measure([0, 1, 2], [0, 1, 2])  # Measure for result verification
analyze_circuit(qc2, "Circuit 2")
```
